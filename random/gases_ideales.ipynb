# Gases Ideales

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

%matplotlib nbagg
```

$$PV = nRT$$

$$P = nRT\frac{1}{V}$$


```python
V = np.linspace(0.1,3,1000)
fig, ax = plt.subplots(1,figsize=(10,4))
plt.suptitle('P vs V')

def update_plot(n,T):
    """
    This function is linked to the sliders and it replots the sine wave when the sliders are modified
    """
    ax.clear()
    R = 0.082 
    P = (n*T*R)/V
    units = 'n = {} moles \nT = {} $K$'
    ax.plot(V,P, label = units.format(n,T))
    ax.legend(loc=1)
    ax.set_xlabel('$L$')
    ax.set_ylabel('atm')
    plt.show()
    
n = widgets.FloatSlider(min= 0.01,max =10, value= 0.1, description = 'n (moles)')
T = widgets.FloatSlider(min= 273,max =500, value= 298 , description = 'T (K)')

widgets.interactive(update_plot,n = n, T = T)
```
